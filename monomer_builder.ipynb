```python
class IDT(mb.Compound):
    def __init__(self):
        super(IDT,self).__init__()
        self.add(mb.load('C1C2=C(C3=CC4=C(C=C31)C5=C(C4)C=CS5)SC=C2',smiles=True))
        carbon1 = list(self.particles_by_name('C'))[0]
        #carbon2 = list(self.particles_by_name('C'))[11]
        carbon3 = list(self.particles_by_name('C'))[14]
        #carbon4 = list(self.particles_by_name('C'))[13]
        self.add(mb.Port(anchor=carbon1, orientation=[0, 1, 4], separation=0.08), "p1")
        #self.add(mb.Port(anchor=carbon1, orientation=[0, 1, 13], separation=0.08), "p2")
        #self.add(mb.Port(anchor=carbon2, orientation=[0, 1, -13], separation=0.08), "p3")
        #self.add(mb.Port(anchor=carbon2, orientation=[0, 1, 13], separation=0.08), "p4")
        self.add(mb.Port(anchor=carbon3, orientation=[0, 1, -13], separation=0.08), "p5")
        #self.add(mb.Port(anchor=carbon4, orientation=[0, 1, -13], separation=0.08), "p6")
        h1 = list(self.particles_by_name('H'))[0]
        h2 = list(self.particles_by_name('H'))[1]
        h3 = list(self.particles_by_name('H'))[4]
        h4 = list(self.particles_by_name('H'))[5]
        h5 = list(self.particles_by_name('H'))[7]
        h6 = list(self.particles_by_name('H'))[8]
        self.remove(h1)
        #self.remove(h2)
        #self.remove(h3)
        #self.remove(h4)
        #self.remove(h5)
        self.remove(h6)
idt = IDT()
idt.visualize(show_ports=True)
```

```python
class BT(mb.Compound):
    def __init__(self):
        super(BT, self).__init__()
        self.add(mb.load("C1=CC2=NSN=C2C=C1", smiles=True))
        #carbon1 = list(self.particles_by_name('C'))[0]
        #self.add(mb.Port(anchor=carbon1, orientation=[-1, 1, 1], separation=0.08), "p1")
        carbon2 = list(self.particles_by_name('C'))[1]
        self.add(mb.Port(anchor=carbon2, orientation=[0, -1, 0], separation=0.08), "p2")
        H1 = list(self.particles_by_name('H'))[0]
        H2 = list(self.particles_by_name('H'))[1]
        #self.remove(H1)
        self.remove(H2)
bt=BT()
bt.visualize(show_ports=True)
```

```python
class R(mb.Compound):
    def __init__(self):
        super(R, self).__init__()
        self.add(mb.load("CCCCCCCCCCCCCCCC", smiles=True))
        carbon1 = list(self.particles_by_name('C'))[0]
        self.add(mb.Port(anchor=carbon1, orientation=[1, 1, 1], separation=0.08), "p1")
        H1 = list(self.particles_by_name('H'))[0]
        self.remove(H1)
r1=R()
r2 = R()
r3 = R()
r4 = R()
r1.visualize(show_ports=True)
```

```python
idt.translate([-1, 0, 0])
r1.translate([0,0,1])
r2.translate([0,0,1])
r3.translate([0,0,1])
r4.translate([0,0,1])

parent = mb.Compound()
parent.add([idt,bt,r1,r2,r3,r4])
parent.visualize(show_ports=True)
```

```python
mb.force_overlap(move_this=bt,
                to_positions=bt['p2'],
                from_positions=idt['p5'])
mb.force_overlap(move_this=r1,
                to_positions=r1['p1'],
                from_positions=idt['p1'])
mb.force_overlap(move_this=r2,
                to_positions=r2['p1'],
                from_positions=idt['p2'])
mb.force_overlap(move_this=r3,
                to_positions=r3['p1'],
                from_positions=idt['p3'])
mb.force_overlap(move_this=r4,
                to_positions=r4['p1'],
                from_positions=idt['p4'])
```

```python
parent.energy_minimize()
parent.visualize(show_ports=False)
```
