```python
import mbuild as mb
import warnings 
import numpy
import ipywidgets as widgets
warnings.filterwarnings('ignore')
```

```python
class CPDT(mb.Compound):
    def __init__(self):
        super(CPDT, self).__init__()
        self.add(mb.load("C1C2=C(SC=C2)C2=C1C=CS2", smiles=True))
        carbon1 = list(self.particles_by_name('C'))[0]
        carbon2 = list(self.particles_by_name('C'))[0]
        carbon3 = list(self.particles_by_name('C'))[8]
        self.add(mb.Port(anchor=carbon1, orientation=[0, -1, 1], separation=0.08), "p1")
        self.add(mb.Port(anchor=carbon2, orientation=[0, -1, -3], separation=0.08), "p2")
        self.add(mb.Port(anchor=carbon3, orientation=[4, 0.5, -1], separation=0.06), "p3")
        hydrogen1 = list(self.particles_by_name('H'))[0]
        hydrogen2 = list(self.particles_by_name('H'))[1]
        hydrogen4 = list(self.particles_by_name('H'))[5]
        self.remove(hydrogen1)
        self.remove(hydrogen2)
        self.remove(hydrogen4)
cpdt1=CPDT()
cpdt1.visualize(show_ports=True)
```

```python
class FBT(mb.Compound):
    def __init__(self):
        super(FBT, self).__init__()
        self.add(mb.load('/Users/madilyn/Projects/repos/Fragment-Library/FBT.mol2'))
        carbon1 = list(self.particles_by_name('C'))[2]
        self.add(mb.Port(anchor=carbon1, orientation=[0, -1, 0], separation=0.08), "p1")
        carbon2 = list(self.particles_by_name('C'))[5]
        #self.add(mb.Port(anchor=carbon2, orientation=[0, 1, 0], separation=0.08), "p2")
        H1 = list(self.particles_by_name('H'))[0]
        H2 = list(self.particles_by_name('H'))[2]
        self.remove(H1)
        #self.remove(H2)
fbt1=FBT()
fbt1.visualize(show_ports=True)
```

```python
#removing the H and adding other terminal port
fbt2=FBT()
H2 = list(fbt2.particles_by_name('H'))[1]
C2 = list(fbt2.particles_by_name('C'))[5]
fbt2.remove(H2)
fbt2.add(mb.Port(anchor=C2,orientation=[0,1,0], separation=0.06), "p2")
fbt2.visualize(show_ports=True)
```

```python
class rgroup(mb.Compound):
    def __init__(self):
        super(rgroup, self).__init__()
        self.add(mb.load('/Users/madilyn/Projects/repos/Fragment-Library/r_groups/C1_rgroup.mol2'))
        carbon = list(self.particles_by_name('C'))[5]
        self.add(mb.Port(anchor=carbon, orientation=[0.005,0,0], separation=0.07), "p1")
        H1 = list(self.particles_by_name('H'))[10]
        self.remove(H1)
r1= rgroup()
r1.visualize(show_ports=True).show()
```

```python
#removing the H and adding other terminal port
cpdt2=CPDT()
hydrogen3 = list(cpdt2.particles_by_name('H'))[0]
carbon4 = list(cpdt2.particles_by_name('C'))[3]
cpdt2.remove(hydrogen3)
cpdt2.add(mb.Port(anchor=carbon4,orientation=[-4, 0.5, 1], separation=0.06), "p4")
cpdt2.visualize(show_ports=True)
```

```python
r2=rgroup()
r3=rgroup()
r4=rgroup()
cpdt1.translate([-1, 0, 0])
cpdt2.translate([-1, 0, 1])
fbt1.translate([-1, 0.5, 0])
fbt2.translate([-1, -0.5, 0])
r1.translate([1, 0, 0])
r2.translate([1, 0, -1])
r3.translate([1, 0, 1])
r4.translate([1, 0, 2])
parent = mb.Compound()
parent.add([cpdt1,cpdt2,fbt1,fbt2,r1,r2,r3,r4])
parent.visualize(show_ports=True)
```

```python
mb.force_overlap(move_this=fbt2,
                 to_positions=cpdt1['p3'],
                 from_positions=fbt2['p2'])
mb.force_overlap(move_this=cpdt2,
                 to_positions=fbt2['p1'],
                 from_positions=cpdt2['p4'])
mb.force_overlap(move_this=fbt1,
                 to_positions=cpdt2['p3'],
                 from_positions=fbt1['p1'])
mb.force_overlap(move_this=r1,
                 to_positions=cpdt1['p1'],
                 from_positions=r1['p1'])
mb.force_overlap(move_this=r2,
                 to_positions=cpdt1['p2'],
                 from_positions=r2['p1'])
mb.force_overlap(move_this=r3,
                 to_positions=cpdt2['p1'],
                 from_positions=r3['p1'])
mb.force_overlap(move_this=r4,
                 to_positions=cpdt2['p2'],
                 from_positions=r4['p1'])
r1.rotate(3.1415/2,[1,0,0])
r2.rotate(3.1415/2,[1,0,0])
parent.visualize(show_ports=False)
```
